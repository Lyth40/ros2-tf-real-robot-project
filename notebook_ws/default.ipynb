For this part of this project, you will create a **ROS2 Service** that detects the **corners** and **obstacles** in the robot arena and broadcast the locations of these features as **Static TFs**. You will create a **ROS2** package that will store your programs.

Follow the steps listed below to get started with this part:

1. Create a **ROS2** package named ***`ros2_tf_project`***:
   - This is the package that will contain the full project.
   - You can create this package as **C++** or **Python** or even both! This is your choice.

2. Create a **Service Server** node that will perform ***feature detection*** using the laser scanner data. You can call this program as ***`feature_detection_service_server`***.

   - You can set the service server name as ***`/feature_detection`***.
   
   - Your program will subscribe to the ***`/scan`*** topic to get the range values from the laser scanner. Make use of the frontal scan ranges to identify and detect the feature that has been requested in the service request. The frontal angle range is your choice - you can choose any value between ***`40°`*** and ***`120°`***, as long as you have the same amount of scan values on either side of the front scan ray.
   
   - You need to come up with your own algorithm idea to detect the positions of **corners** and **obstacles** in the arena.
   
   - Initially, you will get the position of the detected feature with respect to the laser scanner frame. Next, you should get the position of this feature with respect to the ***`odom`*** frame using the **TF Lookup** technique. Finally, you will broadcast the position of this feature with respect to the ***`odom`*** frame as a **Static TF**.
   
   - In addition to broadcasting the positions of all the corners and obstacles, locate and broadcast the starting position of the robot also as a **Static TF**.
   
   - Once you get the position of the features with respect to the ***`odom`*** frame, save the **x, y, z** coordinates in to a **yaml** file. You will use these points in last part of this project.
   
   - You will have to use a custom service interface for your service server. Call this service interface as ***`FeatureDetection.srv`*** and create it within the ***`ros2_tf_project`*** package itself.

For this part of the project, you will create another **ROS2 Service** that will broadcast one **waypoint** frame as **Dynamic TF**. This frame will be moved around the arena following a simple predefined trajectory shape like an ellipse or a rectangle or any suitable shape of your choice.

Follow the steps listed below to get started with this part:

1. Create a **Service Server** node that will perform **simple trajectory** motion to continuously move a **TF** frame. You can call this program as ***`simple_trajectory_service_server`***.

   - You can set the service server name as ***`/simple_trajectory`***.

   - Initially, you will read all the **features** from the **static frames yaml** file and broadcast all of them as **Static TFs**.

   - Then, you will get the ***`x`*** and ***`y`*** coordinates for the center point of the robot arena from the relevant **feature** coordinates. This ***`x`*** and ***`y`*** coordinate values for the center point will be the offset in ***`x`*** and ***`y`*** axes from the actual ***`odom`*** frame which is not located at the center of the arena.

   - Then, you will calculate the ***`x`*** and ***`y`*** coordinates for the desired trajectory shape in which your **waypoint** frame will move, based on the center values that you calculated in the previous step.

   - You will iteratively broadcast the position of the **waypoint** frame with respect to the ***`odom`*** frame to move it around the arena as a **Dynamic TF**, based on the calculated center point that is offset from the ***`odom`*** frame.

   - Once the **waypoint** frame goes around the arena and comes back to the starting position, that is, after making one full lap around the arena, you will stop broadcasting the **waypoint** frame.

   - Finally, you will complete the service and finalize the service response.

   - You will, again, have to use a custom service interface for your service server. Call this service interface as ***`SimpleTrajectory.srv`*** and create it within the ***`ros2_tf_project`*** package itself.

   - **Note:** The robot movement is not required for this part. Only frame movement is the goal of this part.

   - You can set the action server name as ***`/simple_navigation`***.

   - Start the **Simple Trajectory Service Server** by executing the launch command. You should be able to see all the **Static TFs** once the service has initialized (before a service request is sent).

   - Initially you will make a service request to the **Simple Trajectory Service Server** to start broadcasting the **waypoint** frame. For initial programming purposes, you can set the speed of **waypoint** frame movement to a slow speed and later you can make it move faster once your program is ready.

   - Once the **waypoint** frame is published and starts moving (slowly during the programming phase), you will program your robot to follow this moving **waypoint** frame.

   - You will manually bring the robot to your desired starting location that you saved as **Static TF** in the first part of this project, using the ***`teleop`*** program. The robot does not have to be exactly at the start position, just close enough should be fine.

The robot will finally come to a complete stop when the **waypoint** frame finishes one full lap around the arena.

2. The **simple path planning** procedure explained:
   - For the sake of keeping things simple, you will not use any advanced navigation algorithms to solve this task.
   - Start by looking up the **TF** between the robot and the odometry frame.
   - Then lookup the **TF** between the odometry frame and the waypoint frame.
   - Now solve for the **distance** between the robot and the waypoint using the information acquired from the above two steps.
   - Then solve for the **angle** between the current robot orientation and the waypoint location, again using the information acquired from the **TF** lookups.
   - Once you have the **distance** and **angle** measures to the waypoint, drive the robot towards the waypoint location from the current position.
   - Make use of suitable linear and angular speed gain multipliers to control the speeds of the robot.
   - Repeat this process iteratively until the robot has completed a full lap around the arena.

Insufficient or inaccurate hardware is the root cause for unreliable robot odometry. This is one of the main reasons why we cannot solely rely on odometry alone for robot navigation.

This is why we use mapping and localization algorithms to help improve the accuracy of the robot location with respect to the world. Sensor fusion algorithms come into play to minimize the position error in localizing a robot in its surrounding environment with the aid of environmental heuristics. You can learn more about mapping, localization and path planning in the **ROS2 Navigation** course that we offer here at **The Construct**.
